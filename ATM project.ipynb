```python
class Atm:
    def __init__(self):
        self.pin=''
        self.balance=0
        self.menu()
    def menu(self):
        print(""" how can i help you 
        1.create pin
        2.change pin 
        3.withdraw
        4.Deposite
        5.Check Balance
        6.Exit
        """)
        a=input("enter your choice")
        if a=='1':      
            self.create_pin()
            self.menu()
        elif a=='2':
            self.change_pin()
            self.menu()
        elif a=='3':
            self.withdraw()
            self.menu()
        elif a=='4':
            self.deposite()
            self.menu()
        elif a=='5':
            self.check_balance()
            self.menu()
        elif a=='6':
            print("Thank you")
        else:
            print("Sorry")
            
    def create_pin(self):
        self.pin=input("entyer your pin")
    def change_pin(self):
        old_pin=input("enter your old pin")
        if old_pin==self.pin:
            self.pin=input("enter your new pin")
        else:
            print("wrong pin")
    def withdraw(self):
        old_pin=input("enter your old pin")
        if old_pin==self.pin:
            if self.balance>0:
                withdraw=int(input("enter amount"))
                if withdraw < self.balance:
                    print(self.balance-withdraw,"withdraw")
                    print("new balance is",self.balance)
                else:
                    print("low balance")
            else:
                print("low balance")
        else:
            print("wrong pin")
    def deposite(self):
        old_pin=input("enter your old pin")
        if old_pin==self.pin:
            deposite=int(input("enter amount"))
            self.balance=self.balance+deposite
            print("your current balance is =",self.balance)
        else:
            print("wrong pin")
    def check_balance(self):
        old_pin=input("enter your old pin")
        if old_pin==self.pin:
            print("your current balance is =",self.balance)
        else:
            print("wrong pin")
```
